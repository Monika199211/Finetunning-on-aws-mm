```python
import sagemaker
from sagemaker.huggingface import HuggingFaceModel

role = sagemaker.get_execution_role()
```

```python
model = HuggingFaceModel(
    model_data="s3://llm-model-artifacts-monika/models/huggingface-pytorch-training-2025-11-15-11-42-14-536/output/model.tar.gz",
    role=role,
    entry_point="inference.py",
    source_dir="inference",
    transformers_version="4.37.0",
    pytorch_version="2.1.0",
    py_version="py310",
)
```

```python
predictor = model.deploy(
    initial_instance_count=1,
    instance_type="ml.g5.xlarge",
    endpoint_name="live-finetune-endpoint"
)
```

```python
print(predictor.predict({"inputs": "what Clinical trials have shown that adding Ezetimibe to statin therapy?"})
```
